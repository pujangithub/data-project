```python
df['Close_log']=np.log(df['Close'])
```

```python
df.Date=pd.to_datetime(df.Date)

df['year']=df['Date'].dt.year
df['month']=df['Date'].dt.month
df['day']=df['Date'].dt.day
```

```python
df.head()
```

```python

features1=['High','Low','Open','Close']
features=['High_log','Low_log','Open_log','Close_log']
plt.subplots(figsize=(20,10))
for i, col in enumerate(features1):
    plt.subplot(2,2,i+1)
    sn.barplot(x='year',y=col,data=df)
plt.show()
```

```python
df['Price_range']= df.High_log-df.Low_log
df['Momentum']=df.Close_log-df.Open_log

# 3. Is it end of month — last 3 days of any month
df['is_month_end'] = np.where(df['day'] >= 28, 1, 0)

# 4. Is it mid month
df['is_mid_month'] = np.where((df['day'] >= 14) & (df['day'] <= 16), 1, 0)
```

```python
df.head()
```

```python
# If tomorrow's Close is higher than today's Close → BUY (1)
# If tomorrow's Close is lower than today's Close → SELL (0)
df['target'] = np.where(df['Close_log'].shift(-1) > df['Close_log'], 1, 0)

print(df['target'].value_counts())
```

```python
plt.figure(figsize=(6,6))
plt.pie(df['target'].value_counts().values,
        labels=["sell (0)","buy (1)"],
        autopct='%1.2f%%',
        colors=['red','blue'])
plt.title("Buy vs Sell Distribution")
plt.show()

```

```python
df1=df.copy()
df1.drop(['Name','Symbol','Date','Open','High','Low','Close'],axis=1, inplace=True)
```

```python
plt.figure(figsize=(12, 10))
sn.heatmap(df1.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df1.drop(['SNo','Open_log','High_log','Low_log'], axis=1, inplace=True)
df1.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X=df1.drop('target', axis=1)
y=df1.target

print(X.shape)
print(y.shape)
```

```python
scaler=StandardScaler()
xnew=scaler.fit_transform(X)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xnew,y,test_size=0.2,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

models={
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=41),
    'SVM' :SVC(kernel='rbf', probability=True)
}

for name , model in models.items():
    model.fit(xtrain,ytrain)
    pred=model.predict(xtest)
    acc=accuracy_score(ytest,pred)

    print(f'{name} Accuracy: {acc:.4f}')
```

```python
for name , model in models.items():
    model.fit(xtrain,ytrain)
    pred=model.predict(xtest)
    print(f'{name}')
    print(classification_report(ytest,pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

plt.figure(figsize=(12,10))
ConfusionMatrixDisplay.from_estimator(
    models["Logistic Regression"],
    xtest,ytest,
    cmap='Blues'
)
plt.show()
```

```python
plt.figure(figsize=(12,10))
ConfusionMatrixDisplay.from_estimator(
    models["Random Forest"],
    xtest,ytest,
    cmap='Blues'
)
plt.show()
```

```python
plt.figure(figsize=(12,10))
ConfusionMatrixDisplay.from_estimator(
    models["SVM"],
    xtest,ytest,
    cmap='Blues'
)
plt.show()
```

```python
import pickle

with open('bitcoin_model.pkl','wb') as f:
    pickle.dump(models["Random Forest"] ,f)

with open ("scaler.pkl" ,'wb') as f:
    pickle.dump(scaler,f)

print("Model saved successfully!")
```
